##**Grouping Data**

```python
import pandas as pd

df= pd.read_csv('/content/drive/MyDrive/Colab Dataset/weather_by_cities.csv')
df
```

```python
g=df.groupby("city")
g
```

```python
for city,data in g:
  print('city:',city)
  print('max:',data.temperature.max())
```

```python
def grouper(df,idx,col):
  if 80<=df[col].loc[idx]<=90:
    return '80-90'
  elif 50 <=df[col].loc[idx]<=60:
    return '50-60'
  else:
    return 'others'
```

```python
g=df.groupby(lambda idx: grouper(df,idx,'temperature'))
g
```

```python
for key, d in g:
  print('Key:',key)
  print('Data:',d)
```
