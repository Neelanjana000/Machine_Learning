##**Data Concatenation and Merging**

```python
import pandas as pd
```

```python
india_weather= pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather= pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
df_n=pd.concat([india_weather,us_weather], ignore_index=True)
df_n
```

```python
df=pd.concat([india_weather,us_weather], keys=['india','us'])
df
```

```python
df.loc['us']
```

```python
temperature_df= pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30]
},index=[0,1,2])
temperature_df
```

```python
windspeed_df= pd.DataFrame({
    "city":["delhi","mumbai"],
    "windspeed":[7,12],
},index=[1,0])
windspeed_df
```

```python
pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
df1= pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
})
df1
```

```python
df2= pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidity":[68,65,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38],
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","san francisco"],
    "humidity":[65,68,71],
})
df2
```

```python
pd.merge(df1,df2,on="city",how='inner')
```

```python
pd.merge(df1,df2,on="city",how='left')
```

```python
pd.merge(df1,df2,on="city",how='right')
```

```python
pd.merge(df1,df2,on="city",how='outer')
```
