##**Data Visualization**

```python
import pandas as pd

df_sales = pd.read_excel('/content/drive/MyDrive/Colab Dataset/linechart.xlsx')
df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'], df_sales['Fridge'],color='blue',label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'],color='green',label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'],color='orange',label='Washing Machine')

plt.title('Product Sales')
plt.xlabel('Financial Quarter')
plt.ylabel('Revenue (mmln $)')
plt.legend()
plt.show()
```

```python
total_sales=df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct='%1.2f%%',explode=(0.1,0,0),shadow=True,startangle=90)
plt.title('Product Sales'
plt.show()
```

```python
df_sales.plot(kind='bar',x='Quarter')
plt.xticks(rotation=45)
plt.show()
```
