###Excel File

```python
df_movies=pd.read_excel('/content/drive/MyDrive/Colab Dataset/movies_db.xlsx','movies')
df_movies
```

```python
df_financials=pd.read_excel('/content/drive/MyDrive/Colab Dataset/movies_db.xlsx','financials')
df_financials.head(4)
```

```python
def standardize_currency(value):
  if value=='$$'or value=='Dollars':
    return 'USD'
  return value
df_financials=pd.read_excel('/content/drive/MyDrive/Colab Dataset/movies_db.xlsx','financials',converters={
    'currency':standardize_currency
})
df_financials.head(5)
```

```python
df_merged=pd.merge(df_movies,df_financials,on='movie_id')
df_merged
```

```python
df_merged.to_excel('movies_financials.xlsx',sheet_name='financials',index=False)
```

```python
df_stocks=pd.DataFrame({
    'tickers':['GOOGL','WMT','MSFT'],
    'price':[845,65,64],
    'pe':[30.37,14.26,30.97],
    'eps':[27.82,4.61,2.12]
})
df_stocks
```

```python
df_weather=pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'event':['Rain','Sunny','Snow']
})
df_weather
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
  df_stocks.to_excel(writer,sheet_name='stocks')
  df_weather.to_excel(writer,sheet_name='weather')
```
